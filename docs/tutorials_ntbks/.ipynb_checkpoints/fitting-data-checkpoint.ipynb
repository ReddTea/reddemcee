# Fitting a model to data
We are going to fit a simple sinusoid model, with some white noise. This model has 5 parameters, 3 for the sinusoid, the offset plus the jitter related to the white noise:

```python
# True parameters
K_ = 55
P_ = 4.1
phase_ = 2.1
offset_ = -5

j_ = 10  # white noise
ndim_ = 5

theta_ = [K_, P_, phase_, offset_, j_]

# We generate the observation time stamps
N_ = 50
baseline = 50
x_ = np.sort(np.random.uniform(0, baseline, N_))

# And their values
# Sinusoid
def sinusoid(theta):
    K, P, pha, offset = theta[:4]
    freq = 2. * np.pi / P
    y = K * np.cos(freq*x_ + pha) + offset
    return y
    
y_ = sinusoid(theta_)
```

Now we need to add some error to our observations:

```python
# we assume a measured error of
yerr_ = abs(np.random.normal(0, j_, N_))

# and we add an error produced by the same procedure to our observations:
y_ += np.random.normal(0, j_, N_)
```

```python
# data
pl.errorbar(x_, y_, yerr=yerr_, fmt=".k", capsize=0)

# true model
x0 = np.linspace(0, baseline, 500)
y0 = K_ * np.cos((2. * np.pi / P_)*x0 + phase_) + offset_

pl.plot(x0, y0, "k", alpha=0.3, lw=3)
pl.xlabel("x")
pl.ylabel("y");
```
