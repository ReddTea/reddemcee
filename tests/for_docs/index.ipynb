```python
import numpy as np
import reddemcee
```

```python
def log_like(x, ivar):
    return -0.5 * np.sum(ivar * x ** 2)

def log_prior(x):
    return 0.0
```

```python
ndim = 2
ntemps, nwalkers, nsweeps, nsteps = 5, 50, 100, 2
```

```python
ivar = 1. / np.random.rand(ndim)
p0 = list(np.random.randn(ntemps, nwalkers, ndim))
```

```python
sampler = reddemcee.PTSampler(nwalkers,
                             ndim,
                             log_like,
                             log_prior,
                             ntemps=ntemps,
                             logl_args=[ivar],
                             )

sampler.run_mcmc(p0, nsweeps, nsteps)
```
